# Nonparametric estimator: Exampirical distribution function

Assume that a r.v $X$ has a distribution function $F(X)$ and that a sample of identically independent distributed (i.i.d) observations $x_1, x_2, \ldots , x_n$ is available. From this sample we will construct an empirical distribution function $F(X|x_1,\ldots , x_n)$. 

For example, if we are measuring the value of resistors from a lot and we get the following values 9.7kΩ, 9.8kΩ, 9.9kΩ, 9.9kΩ, 10.1 kΩ, 10.1kΩ, 10.4kΩ, 10.4kΩ, 10.7kΩ, 10.9kΩ, what is the probability of selecting a resitor less or equal to 10.75kΩ, that is $F(X = 10.75)$

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size': 20})


# [ 9.7kΩ, 9.8kΩ, 9.9kΩ, 9.9kΩ, 10.1 kΩ, 10.1kΩ, 10.4kΩ, 10.4kΩ, 10.7kΩ, 10.9kΩ]
observations = np.array([ 9.7, 9.8, 9.9, 9.9, 10.1, 10.1, 10.4, 10.4, 10.7, 10.9])
n = len(observations)

F_X_10p75 = 1.0*len(observations[observations < 10.75])/n # the 1.0 is for turning the numerator into type double

print 'Probability of selecting a resitor with value 10.75kΩ or less is {}'.format(F_X_10p75)
```

```python
# now let's can build the empirical distribution function for all values of x

F_X = [] 
for x in observations:
    F_X.append(1.0*len(observations[observations <= x])/n)
    
fig = plt.figure(figsize=(10, 5))
plt.step(observations, F_X, 'k', linewidth=2)
plt.xlabel('X')
plt.ylabel('F(X)')
plt.grid()
plt.show()
```

```python
# From F(X) we can calculate P(X) since it is its derivative

p_X = np.diff(F_X)

fig = plt.figure(figsize=(10, 5))
plt.stem(observations[:-1], p_X, 'k', linewidth=2)
plt.xlabel('X')
plt.ylabel('p(X)')
plt.grid()
plt.show()
```
