#Estimating parameters with Bayes (Bernulli distribution)


Suppose, naively, that you are unsure about the probability of heads in a coin flip (spoiler alert: it's 50%). You believe there is some true underlying ratio, call it $p$, but have no previous knowledge or prior opinion on what $p$ might be.

We begin to flip a coin, and record the observations: either HH or TT. This is our observed data. An interesting question to ask is how our inference changes as we observe more and more data? More specifically, what do our posterior probabilities look like when we have little data, versus when we have lots of data.

*Source: Davidson-Pilon, Cam. "Probabilistic programming and bayesian methods for hackers." (2014).

## Binomial trials

We start by identifying that flipping a coin is follows a Binomial distribution defined as:

$$P(X|n,k,p) = \binom{n}{k}p^{n}(1-p)^{n-k}$$

where $n$ is the number of experiment, $k$ the number of expected success, and $p$ the success ratio. Since we are interested in finding the parameters given the observations we recorded, then we are intereted in $P(n,k,p|X)$, which from Bayes is computed as: 

$$P(n,k,p|X) = \frac{P(X|n,k,p)P(n,k,p)}{P(X)}$$

for the coin experiment we have the number of observation, that is, the value of $n$, and $k=1$ (a Bernulli trial). We need then to estimate $p$ using above equation. $P(X)$ is the normalization constant and therefore we do not need to computed to estimate the value of $p$:

$$P(p|X) \propto P(X|p)P(p)$$

We still need to specify the model for the prior probability $P(p)$ since we already have $P(X|p)$, which is the binomial distribution. A convenient choice for the prior is to use the conjugate distribution of the binomial distribution, because it gives us a close solucion for the posterior $P(p|X)$, which will be the same family as the distribution of the prior. The conjugate prior for the binomial is the beta distribution 

$$P_{beta}(x|\alpha, \beta) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{\int_0^1 u^{\alpha-1} (1-u)^{\beta-1}\, du} = \frac{1}{B(\alpha,\beta)} x^{\alpha-1}(1-x)^{\beta-1}$$

Solving for the Posterior $P(p|X)$ with a beta prior results in another prior with parameters $P_{beta}(X|\alpha + s, \beta + f)$. For the proof: https://en.wikipedia.org/wiki/Conjugate_prior#Example

## Estimacion de $p$ usando la funcion prior conjugada 

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats

plt.rcParams.update({'font.size': 20})
```

```python
# Aqui esta la funcion conjugada de la binomial = beta
dist = stats.beta

# Probemos con diferente tamaño de experimentos para ver como cambia la estimacion de p
n_trials = [0, 1, 2, 3, 4, 5, 8, 15, 50, 500]

# Aqui simulamos el lanzamiento de la moneda
data = stats.bernoulli.rvs(0.5, size=n_trials[-1])
x = np.linspace(0, 1, 100)

# Preparamos la figura
fig = plt.figure(figsize=(20, 20))
# Y para cada uno de los experimentos
for k, N in enumerate(n_trials):
     
    heads = data[:N].sum() # Numero de caras: exitos del experimento 
    #calculamos la prob. posterior P(p|x1,..xn) como la Beta(1 + caras,1 + sellos),
    y = dist.pdf(x, 1 + heads, 1 + N - heads)    
    
    #La figura de la prob. posterior
    sx = plt.subplot(len(n_trials) / 2, 2, k + 1)
    plt.xlabel("$p$, probability of heads")
    plt.setp(sx.get_yticklabels(), visible=False)   
    plt.plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)
    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)

plt.suptitle("Bayesian updating of posterior probabilities",y=1.02,fontsize=14)
plt.tight_layout()
```

# Estimacion de $p$ usando Bayes recursivamente

```python
tosses = 5 # numero de lanzamientos de la moneda en un solo experimento
prior = [1./len(x) for e in x] # probabilidad prior, uniforme
iterations = 100 # numero de repeticiones del experimento

fig = plt.figure(figsize=(20, 20))
for e in range(iterations):
    
    # simulamos el lanzamiento de la moneda con una dist. bernoulli
    data = stats.bernoulli.rvs(0.5, size=tosses)    
    
    heads = data.sum() # numero de exitos
    likelihood = stats.binom.pmf(heads, tosses, x) # likelihood es una dist. binomial
    posterior = likelihood * prior 
    normalization = 1./sum(posterior) # normalizacion
    
    if e%(iterations/10) == 0:
        sx = plt.subplot(5, 2, e/10+1)
        plt.xlabel("$p$, probability of heads")
        plt.setp(sx.get_yticklabels(), visible=False)   
        plt.plot(x,likelihood/max(likelihood)*max(prior), 'red',label='likelihood')
        plt.plot(x,prior, 'blue',label='posterior')
        plt.axis([0, 1, 0, 1.1*max(prior)])
        plt.fill_between(x, 0, prior, color="#348ABD", alpha=0.4)
        leg = plt.legend()        
        
    prior = normalization * posterior    
```
