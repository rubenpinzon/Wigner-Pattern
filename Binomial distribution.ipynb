#Example of a Binomial distribution

This shows an example of a binomial distribution with various parameters. Source: http://www.astroml.org/book_figures/chapter3/fig_binomial_distribution.html#book-fig-chapter3-fig-binomial-distribution


```python
import numpy as np
from scipy.stats import binom
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
n_values = [20, 20, 40]
b_values = [0.2, 0.6, 0.6]
linestyles = ['-', '--', ':']
x = np.arange(-1, 200)

#------------------------------------------------------------
# plot the distributions
fig, ax = plt.subplots(2, figsize=(10, 10))

for (n, b, ls) in zip(n_values, b_values, linestyles):
    # create a binomial distribution
    dist = binom(n, b)

    ax[0].plot(x, dist.pmf(x), ls=ls,
             label=r'$b=%.1f,\ n=%i$' % (b, n), linestyle='steps-mid')  
    ax[1].plot(x, dist.cdf(x), ls=ls,
             label=r'$b=%.1f,\ n=%i$' % (b, n), linestyle='steps-mid') 
    
#for the density function
ax[0].set_xlim(-0.5, 35)
ax[0].set_ylim(0, 0.25)
ax[0].set_xlabel('$x$')
ax[0].set_ylabel(r'$p(x|b, n)$')
ax[0].set_title('Binomial Distribution')
ax[0].legend()

#for the cummulative density function
ax[1].set_xlim(-0.5, 35)
ax[1].set_ylim(0, 0.25)
ax[1].set_xlabel('$x$')
ax[1].set_ylabel(r'$F(x|b, n)$')
ax[1].set_title('CDF Binomial Distribution')
ax[1].legend()

plt.show()

```

##For the Bernulli distribution (n=1)

```python
n_observations = 100
p = 0.3 #success rate
Y = np.random.binomial(1, p, n_observations)

fig, ax = plt.subplots(2, figsize=(10, 10))

ax[0].plot(range(n_observations), Y, 'x')
ax[1].hist(-Y, bins=2)

ax[0].yaxis.set(ticks=(0,1), ticklabels=('Failure', 'Success'))
ax[1].xaxis.set(ticks=(-1,0), ticklabels=('Success', 'Failure'));

ax[0].set(title=r'Bernoulli Trial Outcomes $(\theta=0.3)$', xlabel='Trial', ylim=(-0.2, 1.2))
ax[1].set(ylabel='Frequency')

```
