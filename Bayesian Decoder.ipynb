# Bayesian linear decoder

We want to provide a reference frame for the analysis carried out with the GPFA latent variable model of the hippocampal data during alternations task using a traditional method for spike analysis. For this purpose a bayesian decoder of animal position based on spike count vectors will be used as presented in similar works [1, 2, 3]. First, pyramidal cells with stable place fields will be extracted by constructing 2D occupancy firing rate maps in the maze normalized by the time the animal spent in each bin. Each bin is defined as a square of edge length 2 cm. The 2D map is then smoothed with a Gaussian filter (s.d. 8 cm). Place field peak rate is defined as the maximum rate across bins [3]. A cell with a stable place field is selected only if the peak is larger than 3 Hz [3] and unimodal [4]. Interneurons are excluded from the analysis.

*[1] Brown EN, Frank LM, Tang D, Quirk MC, Wilson MA. A statistical paradigm for neural spike train decoding applied to position prediction from ensemble firing patterns of rat hippocampal place cells. J. Neurosci. 1998; 18:7411–7425.*

*[2] Zhang K, Ginzburg I, McNaughton BL, Sejnowski TJ. Interpreting neuronal population activity by reconstruction: unified framework with application to hippocampal place cells. J. Neurophysiol.1998; 79:1017–1044.*

*[3] Mattias PK, Loren MF. Awake replay of remote experiences in the hippocampus. Nat Neurosci. 2009 Jul;12(7):913-8*

*[4] Foster, DJ, Matthew AW. Reverse replay of behavioural sequences in hippocampal place cells during the awake state. Nature 440.7084 (2006): 680-683.*

##Normalized occupancy firing rate

Compute the linear distance travelled by the mouse on each trial, then compute the spike count on bins of 2 cm and proceed with the normalize firing rate, using the time the animal spent on each bin.

```python
plt.figure(num=None)
f, arr = plt.subplots(2,figsize=(10,10),dpi=650)

cumdist = defaultdict(list)
for trial in data: 
    d = trial.cumulative_dist()
    cumdist[trial.type].append(d) 

# Extract the anomalous trials as those with cummulative linear distance beyond mu+/-s.d
for k in ['left', 'right']:
    vs = cumdist[k]
    mx_dist = map(lambda x: len(x), vs)
    cumdist['mu_' + k] = np.mean(mx_dist)
    cumdist['sd_' + k] = np.std(mx_dist)
    
#Disable trials that wont be used
test_cell = 5
max_d = 0
print 'Showing neuron {} in the linearized distance as control'.format(test_cell)
for trial in data: 
    d = trial.cumulative_dist()
    max_d = max(max_d, max(d))
    if trial.type in ['left', 'right']:
        keep = abs(len(d) - cumdist['mu_' + trial.type]) < cumdist['sd_' + trial.type]
        if keep:    
            subp = 0 if trial.type is 'left' else 1
            arr[subp].plot(d, color=trial.color)
            arr[subp].plot(trial.spikes_lin[test_cell][0],trial.spikes_lin[test_cell][1], 'x', color='k')
            
        else:
            trial.disable()     
    else:
        trial.disable()  
      
plt.xlabel('sample points (mm)')
```
