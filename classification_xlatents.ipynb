```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors
from sklearn.neighbors import NearestCentroid
import scipy.io as sio
from sklearn.naive_bayes import GaussianNB
from sklearn import cross_validation
from sklearn.metrics import confusion_matrix

x = sio.loadmat('/home/ruben/Documents/HAS/HC-5/i01_maze13_003/i01_maze13_003x_orth.mat')
x_orth = np.transpose(x['x_knn'],[2, 1, 0])
y = np.concatenate(x['label'])
```

```python
%matplotlib inline

def bayesClass(x, y, verbose=True):
    n_folds = 5
    acc = []
    [n_laps, n_bins, n_lat] = x_orth.shape
    for t in range(n_bins):
        X = np.zeros([n_laps, n_lat])
        for lap in range(n_laps):
            X[lap,:] = x_orth[lap,t,:]

        acc_within_fold = []
        for k in range(n_folds):
            X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.3, random_state=k)
            gnb = GaussianNB()
            y_pred = gnb.fit(X_train, y_train).predict(X_test)
            #print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0],(y_test != y_pred).sum()))
            cf = confusion_matrix(y_test, y_pred)
            acc_within_fold.append((cf[0,0] + cf[1,1]))
        if verbose:
            print("Bin %d = Accuracy (TP+TN/(P+N)) = %1.2f+/-%1.2f%%" % (t, np.mean(acc_within_fold)/X_test.shape[0],np.std(acc_within_fold)/X_test.shape[0]))
        acc.append((np.mean(acc_within_fold)/X_test.shape[0],np.std(acc_within_fold)/X_test.shape[0]))
        
    mu = [x[0] for x in acc]
    yerr = [x[1] for x in acc]
    tbins = range(n_bins)

    upper_sd = [m+s for m,s in zip(mu,yerr)]
    lower_sd = [m-s for m,s in zip(mu,yerr)]


    # plot it!
    fig, ax = plt.subplots(1)
    ax.plot(tbins, mu, lw=2, label='Acc', color='blue')
    ax.fill_between(tbins, upper_sd, lower_sd, facecolor='blue', alpha=0.5)
    ax.grid()
    ax.set_title('Classification of each time point')
    ax.set_xlabel('Bins')
    ax.set_ylabel('Classification Acc')

    print ("Overall Accuracy (TP+TN/(P+N)) = %1.2f+/-%1.2f%%" % (np.mean(mu), np.max(yerr)))
```

```python
bayesClass(x, y, verbose=False)
```
